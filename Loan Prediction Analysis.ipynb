```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('train.csv')
df
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df['Credit_History'].replace(np.nan,1.0,inplace=True)
```

```python
a=df['Loan_Amount_Term'].mode()
df['Loan_Amount_Term'].replace(np.nan,float(a),inplace=True)
```

```python
df.isna().sum()
```

```python
df['LoanAmount'].replace(np.nan,df['LoanAmount'].mean(),inplace=True)
```

```python
df.isna().sum()
```

```python
df.dropna(subset=['Married'],inplace=True)
```

```python
b=df['Gender'].mode()
df['Gender'].replace(np.nan,str(b),inplace=True)
```

```python
df.isna().sum()
```

```python
c=df['Dependents'].mode()
df['Dependents'].replace(np.nan,str(c),inplace=True)
```

```python
df.isna().sum()
```

```python
d=df['Self_Employed'].mode()
df['Self_Employed'].replace(np.nan,str(d),inplace=True)
```

```python
df.isna().sum()
```

```python
df.info()
```

```python
x=df.drop(columns='Loan_Status')
y=df['Loan_Status']
```

```python
y
```

```python
from sklearn.preprocessing import LabelEncoder
val=LabelEncoder()
x['Loan_ID']=val.fit_transform(x['Loan_ID'])
x['Gender']=val.fit_transform(x['Gender'])
x['Married']=val.fit_transform(x['Married'])
x['Dependents']=val.fit_transform(x['Dependents'])
x['Education']=val.fit_transform(x['Education'])
x['Self_Employed']=val.fit_transform(x['Self_Employed'])
x['Property_Area']=val.fit_transform(x['Property_Area'])
x
```

```python
from sklearn.preprocessing import LabelEncoder
val=LabelEncoder()
y=val.fit_transform(y)
df2=pd.DataFrame(y)
df2.columns=['Loan_Status']
df2
```

```python
x.corr()
```

```python
def correlation(dataset, threshold):
    col_corr = set()
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold: 
                colname = corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
```

```python
corrl=correlation(x,0.7)
corrl
```

```python
x.drop(corrl,axis=1,inplace=True)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,df2,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression  
regressor= LogisticRegression()  
regressor.fit(x_train, y_train)  
```

```python
x_testpred=regressor.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(x_testpred,y_test)
```
